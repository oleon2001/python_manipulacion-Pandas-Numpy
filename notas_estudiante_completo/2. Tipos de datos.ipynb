```
import numpy as np
```

```
matriz = np.array([1,2,3])
matriz.dtype
```

```
matriz = np.array([1,2,3], dtype= 'int32')
matriz.dtype
```

```
matriz = np.array([1,2,3], dtype = 'float32')
matriz.dtype
matriz
```

```
matriz = np.array([1,2,3])
matriz = matriz.astype(np.bool_)
matriz.dtype
```
