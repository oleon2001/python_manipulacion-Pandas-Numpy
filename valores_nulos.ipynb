### Importar librerías

```python
import pandas as pd 
import numpy as np 
```

### Crear diccionario y DataFrame

```python
dict = {'Col1':[1,2,3,np.nan],
 'Col2':[4,np.nan,6,7],
 'Col3':['a','b','c',None]}
dict
```

```python
df = pd.DataFrame(dict)
```

```python
df
```

### Ver valores nulos

```python
df.isnull() # evalua si hay valores que son nulos tanto strings como numeros
```

```python
df.isnull()*1 # a los valores nulos les realizas operaciones
```

### Llenar valores nulos

```python
df.fillna('Missing') #se le asigna un valor si encuentra un valor nulo
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols]) #  Rellena los valores NaN en cada columna numérica con la media de esa columna.
df
```

```python
non_numeric_cols = df.select_dtypes(exclude=['number']).columns
df[non_numeric_cols] = df[non_numeric_cols].fillna('')
df
```

### Interpolar

```python
df.interpolate() #toma los valores NULL NUMERICOS  y les da un valor siguiendo una serie
```

### Borrar valores nulos

```python
df.dropna() # aqui borra los valores nulos 
```
