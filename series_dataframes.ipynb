### Crear DataFrame

```python
dict = {'Jugador':['Navas','Mbappe','Neymar','Messi'],
 'Altura':[183.0, 170.0, 170.0, 165.0],
 'Goles':[2,200,200,200]
 }
```

```python
df_Players = pd.DataFrame(dict,index=[1,7,10,30]) #no se utiliza el series porque estamos hablando de una estructura de mas de una dimension; hay 3 dimesiones distintas
# se agregan los indices de una vez dentro de la invocacion de la libreria 
df_Players
```
