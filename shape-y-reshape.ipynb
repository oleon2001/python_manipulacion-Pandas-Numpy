### Reshape

```python
arr.reshape(1,6) # cambia las filas y las columnas del arreglo en este caso a 1 x 6 en contraposicion del original que es 3 x 2
```

```python
arr.reshape(2,3) #aqui se cambia a 2 x3 
```

```python
np.reshape(arr,(1,6)) # aqui se usa la libreria de numpy es lo mismo 
```

```python
arr
```

```python
np.reshape(arr,(2,3),'C') # aqui toma el reshape y los orderna como lo hace C en este caso por filas 
```

```python
np.reshape(arr,(2,3),'F') # aqui hace lo mismo pero por columnas porque asi lo hace fortran
```

```python
np.reshape(arr,(2,3),'A') # aqui define que los ordene como esta almacenado en el equipo el decide si lo usa de tipo F o C dependiendo de que lea en los datos del equipo y como este lo encuentre mas optimizado mostrar
```

```python
arr = np.random.randint(1,9,(4,6))
```

```python
arr
```

```python
arr.reshape(2,-1)
```

La línea de código arr.reshape(2, -1) se utiliza para cambiar la forma de un array NumPy a una nueva forma especificada. En este caso, la nueva forma es una matriz de 2 filas y un número de columnas que se infiere del tamaño total del array original y la primera dimensión.

```python
arr.reshape(2,2,-1)
```

Dimension -1: El uso de -1 en la tercera dimensión de newshape indica a la función reshape que el número de elementos en esa dimensión debe inferirse del tamaño total del array original y las dos primeras dimensiones. Esto significa que el array se redistribuirá en las 2 filas y 2 columnas de la nueva forma, manteniendo el mismo número total de elementos.
