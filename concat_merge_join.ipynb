### Importar librerías

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3'],})

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7'],})
```

### concat()

```python
pd.concat([df1,df2],ignore_index=True) # se le coloca el atributo de ignore_index=True para ignorar los indices originales
```

```python
pd.concat([df1,df2]) #como aqui se muestra
```

```python
pd.concat([df1,df2],axis=1) # cuando se le coloca el concat en el axis=1 se juntan las matrices en filas y no por columnas como el caso anterior por default
```

### merge()

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
izq.merge(der, on='key')# aqui hace una combinacion de los dataframes en los cuales siempre los que estan entre parentesis se posicionara a la derecha del dataframe
# el atributo on='key' sirve para fusionar en este caso los keys para mantener la congruencia mas debido a una version de pandas ya esto no es necesario mas seria una buena practica
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
izq.merge(der, left_on='key', right_on='key_2') 
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2',np.nan],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
der
```

### how

```python
izq.merge(der, left_on='key', right_on='key_2', how='right') # el how se agrega para determinar el tipo de union ya sea un left, right, inner o outer join 
```

### join() - Index match

```python
izq = pd.DataFrame({'A':['A0','A1','A2'],
 'B':['B0','B1','B2']},
 index=['k0','k1','k2'])

der = pd.DataFrame({'C':['C0','C1','C2'],
 'D':['D0','D1','D2']},
 index=['k0','k2','k3'])
```

```python
izq.join(der,how='outer')
```
