### Importar librerías

```python
import pandas as pd
import numpy as np 
```

### Cargar DataFrame

```python
df_books = pd.read_csv('bestsellers-with-categories.csv',sep=',',header=0)
```

```python
df_books
```

```python
df_books.head(2) # retorna las 2 primeras filas
```

### Drop columns

```python
df_books.drop('Genre',axis=1,inplace=True) #axis es el eje esto por si solo no lo borra del dataframe si no de la salida para borralo hay que agregar el atributo inplace=true
```

```python
df_books = df_books.drop('Year',axis=1) #esta es otra forma de realizarlo para almacenar los valores en el dataframe
'''
Recordar:

    axis=0: se refiere a las filas.
    axis=1: se refiere a las columnas.


'''
```

```python
df_books.head(2)
```

```python
del df_books['Price'] #propio del lenguaje no se recomienda en estos casos
```

### Drop rows

```python
df_books.head(2)
```

```python
df_books.drop(range(0,10),axis=0).head(2) # el range borra todo los indices que estan desde la posicion 0 hasta la 10 
'''
Recordar:
    axis=0: se refiere a las filas.
    axis=1: se refiere a las columnas.
'''
```

### Add columns

```python
df_books.head(2)
```

```python
df_books['Nueva_Columna'] = np.nan #se importa numpy para agregar una nueva comlumna el np.nan se usa para agregar valores no numericos
df_books.head(2)
```

```python
data = np.arange(0,df_books.shape[0]) # se crean un array desde la posicion 0 hasta el tamaño de los indices que hay en el dataframe 
data
```

### Add rows

```python
df_books = pd.concat([df_books, df_books], ignore_index=True) # asi se concatenan unos dataframes con otros dataframes 
df_books
```
